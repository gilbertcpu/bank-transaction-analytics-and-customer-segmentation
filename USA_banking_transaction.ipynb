```python
# Aggregate dataset to customer level
customer_df = df.groupby('Account_Number').agg({
    'Transaction_Amount': ['sum', 'mean'],
    'Transaction_ID': 'count',
    'Customer_Age': 'first',
    'Customer_Income': 'first',
    'Loyalty_Points_Earned': 'sum'
})

customer_df.columns = ['Total_Spend', 'Avg_Spend', 'Transaction_Count', 'Age', 'Income', 'Total_Loyalty']

customer_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

features = ['Total_Spend', 'Avg_Spend', 'Transaction_Count', 'Age', 'Income', 'Total_Loyalty']

scaler = StandardScaler()
scaled = scaler.fit_transform(customer_df[features])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []
for k in range(2, 10):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(scaled)
    wcss.append(model.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(2,10), wcss, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
customer_df['Cluster'] = kmeans.fit_predict(scaled)
customer_df.head()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
components = pca.fit_transform(scaled)
customer_df['PCA1'] = components[:,0]
customer_df['PCA2'] = components[:,1]

plt.figure(figsize=(8,6))
sns.scatterplot(data=customer_df, x='PCA1', y='PCA2', hue='Cluster', palette='bright')
plt.title("Customer Segmentation (K-Means)")
plt.show()
```

Cluster 0 (Biru)

-Sebagian besar berada di area PCA2 tinggi

-Mengindikasikan aktivitas transaksi lebih tinggi

-Spending cenderung moderate

-Usia produktif & loyalitas stabil

Segment: Active Daily Users

Strategi: dorong cross-selling (kartu debit  kredit, tabungan  investasi)

Cluster 1 (Oranye)

-Tersebar di bagian kanan grafik (PCA1 tinggi)

-Income & spending di atas rata-rata

-Loyalitas masih variasi, belum sangat kuat

Segment: Affluent but Less Engaged

-trategi: loyalty rewards, premium membership, upgrade financial services

Cluster 2 (Hijau)

-Paling terkonsentrasi di kanan atas  high PCA1 & PCA2

-Kategori paling tinggi nilai finansial & aktivitas transaksi

Segment: High-Value & Loyal Customers

Strategi:

-Prioritaskan dengan layanan personal banking

-Special privilege, high credit limit, investment advisory

-Retention sangat penting

Cluster 3 (Merah)

-Dikumpulkan di area bawah grafik (PCA2 rendah, PCA1 rendahmedium)

-Income rendah, spending rendah, transaksi rendah

Segment: Low Engagement / At Risk

Strategi:

-Promosi cashback

-Fee waivers

-Aktivasi ulang layanan digital banking
